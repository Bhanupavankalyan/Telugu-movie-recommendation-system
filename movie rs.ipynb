```python
import pandas as pd
import re

df = pd.read_csv("tollywood_movies_dataset.csv")

def fix_list_format(text):
    if isinstance(text, list):
        return text
    text = str(text)
    # Remove outer spaces
    text = text.strip()

    # Replace unquoted words with quoted words
    text = re.sub(r"(?<=\[)(.*?)(?=\])", lambda m: 
                  ", ".join([f'"{item.strip()}"' for item in m.group(1).split(',')]), text)

    return eval(text)

df['genres'] = df['genres'].apply(fix_list_format)
df['cast'] = df['cast'].apply(fix_list_format)
```

```python
import pandas as pd
import re

df = pd.read_csv("tollywood_movies_dataset.csv")

def fix_list(text):
    text = str(text).strip()

    # Remove [ and ]
    inner = text.strip("[]").strip()

    if not inner:
        return []

    # Split by comma
    parts = [p.strip() for p in inner.split(",")]

    # Add quotes to every word
    parts = [f'"{p}"' for p in parts]

    # Join back into valid list string
    fixed = "[" + ", ".join(parts) + "]"

    # Now safely eval
    return eval(fixed)

df['genres'] = df['genres'].apply(fix_list)
df['cast']   = df['cast'].apply(lambda x: fix_list(x)[:3])
```

```python
# Convert JSON strings to Python lists
import ast

def convert(text):
    try:
        return ast.literal_eval(text)
    except:
        return []

df['genres'] = df['genres'].apply(convert)
df['cast'] = df['cast'].apply(convert)
```

```python
def get_names(list_data):
    return [item['name'] for item in list_data if 'name' in item]

df['genres'] = df['genres'].apply(get_names)
df['cast'] = df['cast'].apply(get_names)
```

```python
df['tags'] = df['genres'] + df['cast'] + df['overview'].apply(lambda x: x.split())
```

```python
df['tags'] = df['tags'].apply(lambda x: " ".join(x))
```

```python
df[['title', 'tags']].head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000, stop_words='english')

vectors = cv.fit_transform(df['tags']).toarray()
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
similarity.shape
```

```python
similarity.shape
```

```python
def recommend(movie):
    movie = movie.lower()

    # Check if movie exists
    if movie not in df['title'].str.lower().values:
        return ["Movie not found in dataset"]

    # Get index of the movie
    idx = df[df['title'].str.lower() == movie].index[0]

    # Get similarity scores
    distances = list(enumerate(similarity[idx]))

    # Sort movies by similarity
    movies_list = sorted(distances, reverse=True, key=lambda x: x[1])[1:6]

    # Return top 5 recommendations
    recommendations = []
    for i in movies_list:
        recommendations.append(df.iloc[i[0]].title)

    return recommendations
```
